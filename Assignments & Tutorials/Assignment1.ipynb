```python
import networkx as nx
import pandas as pd
import numpy as np
from networkx.algorithms import bipartite
%matplotlib notebook


# Open the text file and read all the employees' names 
file = open('Employee_Movie_Choices.txt','r')
file_data = file.readlines()

employee_list = []
movie_list = []
unique_employee_list = []
unique_movie_list = []

for line in file_data:
    word = line.split("\t")
    word[1] = word[1][:-1]          #remove \n
    employee_list.append(word[0])
    movie_list.append(word[1])
    

# Delete the column header
del employee_list[0]   
del movie_list[0]

# List & Set
employee_set = set(employee_list)
unique_employee_list = list(employee_set)
movie_set = set(movie_list)
unique_movie_list = list(movie_set)

set_list = []
for i in range(len(employee_list)):
    set_list.append((employee_list[i], movie_list[i]))

# Create the L-Bipartite graph
B = nx.Graph()
B.add_nodes_from(employee_list, bipartite = 0)
B.add_nodes_from(movie_list, bipartite = 1)
B.add_edges_from(set_list, bipartite = 0)
nx.draw_networkx(B, font_size = 8, node_color = '#808080', alpha = 0.5, width = 0.3, with_labels = True )

# Check the node and edge numbers 
print ("Number of nodes and edges in the graph: ")
print (len(B))  # number of nodes in B
print (B.number_of_edges())  # number of edges in B
```
